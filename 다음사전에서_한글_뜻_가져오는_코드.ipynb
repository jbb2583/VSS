```
from urllib.request import urlopen
import pandas as pd
import numpy as np
import bs4
word_list = pd.read_csv('word_List.csv',header=None) #CSV 파일 불러오기
word_list=word_list.values

```

```
for i in range(word_list.size): # 불러온 파일 단어 나열하기
  print(word_list[i,0])
```

```
f = open("word_listing.txt",'w') #Text 파일 하나 생성 및 수정 가능한 상태로 열기
for i in range(word_list.size): # 단어마다 Loop 돌기
  index_cd = word_list[i,0] 
  naver_index = 'https://small.dic.daum.net/search.do?q=' + index_cd # 각 단어에 대해 검색
  source = urlopen(naver_index).read() # 소스 파일 불러오기
  source = bs4.BeautifulSoup(source,'lxml') #bs4 라이브러리 사용
  Data = source.find_all('ul')[2].text #'ul'클래스의 두 번째 것(다음 사전 단어에서의 뜻이 있는 항목) 찾아 슬라이싱 
  Data = str(Data).replace('\n',' ') # 개행 문자 공백으로 변환 
  Data = word_list[i,0] + Data + '\n' #영어 단어 + 한글 뜻 + 개행 문자로 저장
  print(Data) #저장된 단어 프린트
  f.write(Data) #텍스트 파일에 쓰기
f.close() # 텍스트 파일 닫기
```
